```python
d={
    "India":140,
    "China":130,
    "USA":30,
    "Russia":14,
    "UK":7
}
print("sort based on keys:")
print(d)
newd=dict(sorted(d.items()))
print(newd)
newd=dict(sorted(d.items(),reverse=1))
print(newd)

print("based on values")
#descending order
newd=dict(sorted(d.items(),key=lambda x:x[1],reverse=1))
print(newd)
#ascending order
newd=dict(sorted(d.items(),key=lambda x:x[1]))
print(newd)
```

```python
def byprice():
    return dict(sorted(d.items(),key=lambda x:x[1]['price']))
def byqty():
    return dict(sorted(d.items(),key=lambda x:x[1]['qty']))
def byname():
        return dict(sorted(d.items(),key=lambda x:x[1]['name']))
def bynameandprice():
        return dict(sorted(d.items(),key=lambda x:(x[1]['name'],x[1]['price'])))
def bynameandqty():
        return dict(sorted(d.items(),key=lambda x:(x[1]['name'],x[1]['qty'])))
def bypriceandqty():
    return dict(sorted(d.items(),key=lambda x:(x[1]['price'],x[1]['qty'])))
def bprice_qty_name():
    return dict(sorted(d.items(),key=lambda x:(x[1]['price'],x[1]['qty'],x[1]['name'])))
d={}
n=int(input("enter no of entries\n"))
for i in range(n):
    k='I'+str(i+1)
    name=input()
    price=int(input())
    qty=int(input())
    print(f"name:{name}, price:{price}, qty:{qty}")
    d1={"name": name,"price": price ,"qty": qty}
    d[k]=d1

    
print(d)
print("\n 1.By Price \n 2.By Qty \n 3.By name 4.By name and price \n 5.By  name and qty \n 6.By price and qty \n 7.By  combo of price , qty and name \n 8.exit")

while 1:
    ch=int(input("enter choice"))
    match (ch):
         case 1:
             print(byprice())
         case 2:
             print(byqty())
         case 3:
             print(byname())
         case 4:
             print(bynameandprice())
         case 5:
             print(bynameandqty())
         case 6:
             print(bypriceandqty())
         case 7:
             print(byprice_qty_name())
         case 8:
             break
         case _:print("invalid choice")
        
```
