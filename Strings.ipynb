```python
#program to read a line io string and create a new string by converting each of their first letter into capital
s=input()
l=s.split()
for i in range(len(l)):
    l[i]=l[i].capitalize()
res=""
for x in l:
    res+=x
    res+=" "
res=res[:-1]
print(res)
```

```python
#a program to take a list of strings as input and print the string combined with first,middle and last character into single string
l=list(map(str,input().split()))
s1="";s2="";s3="";
for x in l:
    if len(x)>2:
        s1+=x[0]
    if len(x)>2 :
        mid=len(x)//2
        if len(x)%2==0:
            s2+=x[mid-1]
        else:
            s2+=x[mid]
    if len(x)>2:
        s3+=x[-1]
print(s1+s2+s3)

```
