```python
#create a class named Book and add constructor and take inputs as title,author,publisher and price and initialize them as private variables
class Book:
    grandtotal=0
    totalcopies=0
    def __init__(self,title,author,publisher,price):
        self.__title=title
        self.__author=author
        self.__publisher=publisher
        self.__price=price
    def purchase(self):
        self.__noc=int(input("enter no of copies:"))
        self.total=self.__price*self.__noc
        self.discount=0
        if self.__noc<100:
            self.discount=0;
        elif self.__noc>=100 and self.__noc<200:
            self.discount=(self.__price*10)/100
        elif self.__noc>=200 and self.__noc<300:
            self.discount=(self.__price*20)/100
        elif self.__noc>=300 and self.__noc<400:
            self.discount=(self.__price*40)/100
        else:
            self.discount=(self.__price*50)/100
        self.total -=self.discount
        Book.grandtotal+=self.total
        Book.totalcopies+=self.__noc

    def display_details(self):        
        print(f"{self.__title} \t {self.__author} \t {self.__price} \t {self.__noc} \t {self.discount} \t\t { self.total }")

            
            
        
            
        
```

```python
n=int(input("enter no of books:"))
l=[]
for i in range(n):
    title=input("enter title:")
    author=input("author:")
    publisher=input("publisher:")
    price=int(input("price:"))
    b1=Book(title,author,publisher,price)
    b1.purchase()
    l.append(b1)
print("Book name \t Author \t Price \t NOC \t Discount \t Total ")

for b in l:
    print(b.display_details())
print("Grand Total:",Book.grandtotal)
print("Total Copies:",Book.totalcopies)

```
