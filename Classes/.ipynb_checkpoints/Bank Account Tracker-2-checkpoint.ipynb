```python
#to take the name of user and create a bank account using classes for 1 user and act as per the user driven menu
class BankAccount:
    seq=100
    def __init__(self,name):
        self.acno=BankAccount.seq
        self.name=name
        self.balance=0
        BankAccount.seq+=1
    def deposit(self,amount):
        self.balance+=amount
        #return True
    def withdrawl(self,amount):
        if self.balance<amount:
            return False
        else:
            self.balance-=amount
            return True
    def checkBalance(self):
        print(self.balance)
    def printDetails(self):
        print(f"{self.acno} \t {self.name} \t {self.balance}")       

        
        
    
    
```

```python
l=[]
print("1.Create account \n 2.Deposit \n 3.WithDrawl \n 4.check balanace \n 5.display details \n 6.exit")
while True:
    ch=int(input("enter your choice:"))
    match(ch):
        case 1:
            name=input("enter name:")
            b=BankAccount(name)
            l.append(b)
        case 2:
            x=float(input("enter the amount to deposit:"))
            b.deposit(x)
            print("Amount Deposited")
        case 3:
            x=float(input("enter the amount to withdrawl:"))
            v=b.withdrawl(x)
            if v:
                print("the amount withdrawn ")
            else:
                print("amount insufficient")
        case 4:
            b.checkBalance()
        case 5:
            print("The Details as follows:")
            print("Acno \t Name \t Balanace")
            for x in l:
                x.printDetails()
        case 6: 
            print("Exiting...")
            break
        case _:
            print("Invalid option")
l.clear()
    
```
