```python
#to take the name of user and create a bank account using classes for 1 user and act as per the user driven menu
class BankAccount:
    seq=100
    def __init__(self,name):
        self.acno=BankAccount.seq
        self.name=name
        self.balance=0
        BankAccount.seq+=1
    def deposit(self,amount):
        self.balance+=amount
        #return True
    def withdrawl(self,amount):
        if self.balance<amount:
            return False
        else:
            self.balance-=amount
            return True
    def checkBalance(self):
        print(self.balance)
    def printDetails(self):
        print('Account Number:',self.acno)
        print('name:',self.name)
        print('balance:',self.balance)
print("Enter account Details:")
name=input("Enter name:")
b=BankAccount(name)
print("1.Deposit \n 2.WithDrawl \n 3.check balanace \n 4.display details \n 5.exit")
while True:
    ch=int(input("enter your choice:"))
    match(ch):
        case 1:
            x=float(input("enter the amount to deposit:"))
            b.deposit(x)
            print("Amount Deposited")
        case 2:
            x=float(input("enter the amount to withdrawl:"))
            v=b.withdrawl(x)
            if v:
                print("the amount withdrawn ")
            else:
                print("amount insufficient")
        case 3:
            b.checkBalance()
        case 4:
            b.printDetails()
        case 5: 
            print("Exiting...")
            break
        case _:
            print("Invalid option")
    
        
        
    
    
```
