```python
s=list(map(str,input().split()))
d={}
res=[]
for i in range(len(s)):
    l=list(s[i])
    l.sort()
    for j in range(i+1,len(s)):
        l1=list(s[j])
        l1.sort()
        if l==l1:
            if s[i] in d:
                
                d[s[i]]=res
               # s.remove(s[j])
            
            else:
                d[s[i]]=s[j]
                #s.remove(s[j])
    if s[i] not in d:
        res.append([s[i]])
for x,y in d.items():
    li=[]
    li.append(x)
    li.extend(y)
    res.append(li)
print(d)      
print(res)
```
