```python
#len of tuple
print("enter elements :")
l=[]
i=1
while True:
    print(f"enter {i}th element (-1 to exit):")
    x=int(input())
    if(x!=-1):
        l.append(x)
        i+=1
    else:
        break
t=tuple(l)
print(t)
print("length of tuple:",len(t))
    
```

```python
#k min and k max
print("enter the elements of tuple:")
l=list(map(int,input().split()))
t=tuple(l)
sortedtuple=sorted(t)
n=len(t)
print("enter k value:")
k=int(input())
if k>len(t):
    print("invalid")
else:
    mint=sortedtuple[:k]
    maxt=sortedtuple[n-k:]
    print(f"{k} min elements:",mint)
    print(f"{k} max elements:",maxt)
    
```

```python
#element,square,cube
n=int(input("enter n value:"))
l=[]
for i in range(n):
    l.append(int(input(f"enter {i+1}th value")))
l1=[]
for x in l:
    li=[x,x**2,x**3]
    
    l1.append(tuple(li))
print(l1)
    
```

```python
t=(1,5,7,8,10)
print(t)
t1=tuple(i*j for i,j in zip(t,t[1:]))
print(t1)
```

```python
l=list(map(int,input("enter elements:").split()))
t=tuple(l)
print("sum is (built in):",sum(t))
s=0
for x in t:
    s+=x
print("sum is (iteration):",s)
```

```python
m=int(input("enter rows"))
l=[]
for i in range(m):
    li=list(map(int,input(f"enter {i+1} row elements:").split()))
    l.append(tuple(li))
print(l)
k=int(input("enter k:"))
for x in l:
    for y in x:
        if y%k==0:
            print(y,end=" ")
```
